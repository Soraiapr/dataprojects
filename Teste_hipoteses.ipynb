```python
import numpy as np
import pandas as pd
from scipy import stats

def teste_normalidade_dados(dados):
    """
    Realiza o teste de normalidade Shapiro-Wilk.

    Parâmetros: dados

    Retorna:
    - stat: Estatística de teste.
    - p_valor: Valor p associado ao teste.
    """
    stat, p_valor = stats.shapiro(dados)
    return stat, p_valor

def teste_hipotese_media_2amostras(dados1, dados2):
    """
    Realiza o teste de hipótese para a diferença entre as médias de duas amostras independentes.

    Parâmetros:
    - dados1
    - dados2

    Retorna:
    - stat: Estatística de teste.
    - p_valor: Valor p associado ao teste.
    """
    stat, p_valor = stats.ttest_ind(dados1, dados2)
    return stat, p_valor

caminho_arquivo = '/content/drive/MyDrive/MBA Data Science/3. Metodologia científica/Dataset/salary_dataset.csv'
dados = pd.read_csv(caminho_arquivo)

# Separar os dados em duas amostras (homens e mulheres)
dados_homem = dados[dados['Gender'] == 'Male']['Salary']
dados_mulher = dados[dados['Gender'] == 'Female']['Salary']

# Realizar o teste de normalidade para a amostra de homens
stat_normalidade_homem, p_valor_normalidade_homem = teste_normalidade_dados(dados_homem)
print(f'Teste de Normalidade - Amostra de Homens - Estatística: {stat_normalidade_homem}, Valor p: {p_valor_normalidade_homem}')

# Interpretar o resultado do teste de normalidade para a amostra de homens
if p_valor_normalidade_homem > 0.05:
    print("Os dados da amostra de homens parecem seguir uma distribuição normal.")
else:
    print("Os dados da amostra de homens não seguem uma distribuição normal.")

# Realizar o teste de normalidade para a amostra de mulheres
stat_normalidade_mulher, p_valor_normalidade_mulher = teste_normalidade_dados(dados_mulher)
print(f'Teste de Normalidade - Amostra de Mulheres - Estatística: {stat_normalidade_mulher}, Valor p: {p_valor_normalidade_mulher}')

# Interpretar o resultado do teste de normalidade para a amostra de mulheres
if p_valor_normalidade_mulher > 0.05:
    print("Os dados da amostra de mulheres parecem seguir uma distribuição normal.")
else:
    print("Os dados da amostra de mulheres não seguem uma distribuição normal.")

# Realizar o teste de hipótese para a diferença entre as médias de homens e mulheres
stat_hipotese, p_valor_hipotese = teste_hipotese_media_2amostras(dados_homem, dados_mulher)
print(f'Teste de Hipótese - Estatística: {stat_hipotese}, Valor p: {p_valor_hipotese}')

# Interpretar o resultado do teste de hipótese
if p_valor_hipotese > 0.05:
    print("Não há evidências suficientes para rejeitar a hipótese nula de que homens ganham o mesmo que mulheres.")
else:
    print("A hipótese nula de que homens ganham o mesmo que mulheres é rejeitada.")
```

Com base nos resultados  do teste de normalidade, analisando o valor p, é possível perceber que os dados tanto da amostra de homens quanto de mulheres não segue uma distribuição normal. E não há evidências suficientes para rejeitar a hipótese nula de que homens ganham o mesmo que mulheres.
Ou seja, com base nos dados disponíveis e no método estatístico utilizado, não há evidências suficientes para afirmar que existe uma diferença salarial significativa entre homens e mulheres na amostra analisada.
É importante notar que a ausência de evidências suficientes para rejeitar a hipótese nula não prova que homens e mulheres ganham exatamente o mesmo. Pode haver outras variáveis ou nuances que não foram consideradas no teste estatístico.

### **Teste Z paramétrico**

```python
import numpy as np
import pandas as pd
from scipy import stats

def teste_z_parametrico(dados1, dados2):
    """
    Realiza o teste Z paramétrico para comparar médias de duas amostras independentes.

    Retorna:
    - stat: Estatística de teste.
    - p_valor: Valor p associado ao teste.
    """
    stat, p_valor = stats.ttest_ind(dados1, dados2)
    return stat, p_valor

caminho_arquivo = '/content/drive/MyDrive/MBA Data Science/3. Metodologia científica/Dataset/salary_dataset.csv'
dados = pd.read_csv(caminho_arquivo)

# Separar os dados em duas amostras (homens e mulheres)
dados_homem = dados[dados['Gender'] == 'Male']['Salary']
dados_mulher = dados[dados['Gender'] == 'Female']['Salary']

# Realizar o teste Z paramétrico
stat_z, p_valor_z = teste_z_parametrico(dados_homem, dados_mulher)
print(f'Teste Z Paramétrico - Estatística: {stat_z}, Valor p: {p_valor_z}')

# Interpretar o resultado do teste Z paramétrico
if p_valor_z > 0.05:
    print("Não há evidências suficientes para rejeitar a hipótese nula de que homens ganham o mesmo que mulheres.")
else:
    print("A hipótese nula de que homens ganham o mesmo que mulheres é rejeitada.")
```

Com base nos dados analisados e no teste estatístico (teste Z paramétrico) realizado, não foi encontrada uma diferença salarial estatisticamente significativa entre homens e mulheres na amostra.

# Gráfico de barras para contar frequência por cargo - genero

```python
import matplotlib.pyplot as plt

parametro = 'Job Title'

def criar_grafico_barras(dados, parametro, gender=None):
    # Filtra os dados pelo gênero
    if gender:
        dados = dados[dados['Gender'] == gender]

    # Filtra os dados pelo parâmetro que contém "director"
    dados = dados[dados[parametro].str.contains('director', case=False, na=False)]

    # Conta a frequência do parâmetro nos dados
    contagem = dados[parametro].value_counts()

    # Prepara os dados para o gráfico
    valores = contagem.index
    frequencias = contagem.values

    plt.bar(valores, frequencias * 1.5, color='skyblue')
    plt.xlabel(parametro)
    plt.ylabel('Frequência')
    plt.title(f'Frequência de Cargos de Diretoria por Gênero Masculino')
    plt.xticks(rotation=90)
    plt.show()

criar_grafico_barras(dados, parametro, gender='Male')
```
